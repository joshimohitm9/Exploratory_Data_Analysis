```python
def rename(team):
  if team=="Royal Challengers Bangalore":
    return "RCB"
  elif team=="Kings XI Punjab":
    return "KXIP"
  elif team=="Delhi Daredevils":
    return "DD"
  elif team =="Mumbai Indians":
    return "MI"
  elif team=="Rajasthan Royals":
    return "RR"
  elif team=="Chennai Super Kings":
    return "CSK"
```

```python
pd.read_csv("/content/drive/MyDrive/IPL Matches 2008-2020.csv",converters={"team1":rename})
```
