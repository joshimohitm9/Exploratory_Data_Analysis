```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("/content/drive/MyDrive/Diwali Sales Data.csv",encoding="unicode_escape")
```

```python
df["Amount"]=df["Amount"].astype("int")
```

**EDA**

```python
gc=sns.countplot(x="Gender",data=df,width=0.7,palette=["pink","red"])
for bars in gc.containers:
  gc.bar_label(bars)
plt.show()
```

```python
ga=df.groupby(["Gender"],as_index=False)["Amount"].sum()
```

```python
ga
```

```python
ab= sns.barplot(x="Gender",y="Amount",data=ga,palette=["blue","purple"],saturation=0.4,width=0.7)
for bars in ab.containers:
  ab.bar_label(bars)
plt.show()
```

```python
df.columns
```

```python
ag=sns.countplot(x="Age Group",data=df,palette="bright")
for bars in ag.containers:
  ag.bar_label(bars)
```

```python
ag=sns.countplot(x="Age Group",data=df,hue="Gender")
for bars in ag.containers:
  ag.bar_label(bars)
```

```python
df.columns
```

```python
ss=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)
```

```python
ss
```

```python
mj=sns.barplot(x="State",y="Orders",data=ss)
sns.set(rc={"figure.figsize":(28,10)})
plt.grid()
for bars in mj.containers:
  mj.bar_label(bars)
```

```python
ms=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
from matplotlib import rc
sns.barplot(x="Marital_Status",y="Amount",data=ms,hue="Gender")
sns.set(rc={"figure.figsize":(5,5)})
```
