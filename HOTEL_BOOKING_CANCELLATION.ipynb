**Finding and Removing Null Values**

```python
df.isnull().sum()
```

```python
for col in df.columns:
 if df[col].isnull().sum()>15000:
    df.drop([col],axis=1,inplace=True)

df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.shape
```

```python
plt.figure(figsize=(20,25))
for i,col in enumerate(df.describe().columns):
    plt.subplot(5,4,i+1)
    plt.boxplot(df[col])
    plt.title(col)
plt.show()
```

**Z score for removing Outliers**

```python
upper_limit=df["adr"].mean()+ 3*df["adr"].std()
lower_limit=df["adr"].mean()-3*df["adr"].std()
print("Upper Limit :",upper_limit)
print("lower_limit",lower_limit)
```

```python
outliers=df[(df["adr"]>upper_limit) | (df["adr"]< lower_limit)]
outliers.shape
```

```python
df=df[(df["adr"]<upper_limit) & (df["adr"]>lower_limit)]
df.shape
```

```python
df.head()
```

```python
df["family"] = df["adults"] + df["children"] + df["babies"]
df["family"]=df["family"].astype(int)
df["family"]
```

```python
df.tail()
```

```python

family_type = []

for i, row in df.iterrows():
    val = row['family']
    if val == 2:
        family_type.append("Couples")
    elif val > 2:
        family_type.append("Family")
    else:
        family_type.append("Single")

df['family_type'] = family_type
```

```python
df.head().T
```

```python
df.shape
```

```python
df.drop(df["family"],axis=0 , inplace =True)
```

##conclusion





* **Cancellation Trends:** Approximately 50% of hotel bookings are canceled, with a
significant portion (75%) of these cancellations occurring in City Hotels. There has been a decline in cancellations from 2016 to 2017, indicating the effectiveness of strategies to reduce cancellations.

* **Seasonal Trends:** The months of August, September, and November see the highest cancellation rates, with August being the peak month at 37%. These months also coincide with the peak booking periods, suggesting a need for targeted strategies during these times.

* **Guest Preferences:** Couples account for the majority of bookings and cancellations. Offering them extra discounts or facilities could help reduce cancellations and increase bookings. Repeating guests are fewer in number, indicating an opportunity for targeted marketing to increase repeat bookings.

* **Hotel Type Analysis:** City Hotels are the most affected by cancellations, especially during holiday months like April, May, and June. This suggests that changes in holiday plans could be a significant factor in cancellations.

* **Recommendations:** Providing extra discounts or facilities to couples, especially during peak booking months, could help reduce cancellations. Targeted marketing strategies aimed at increasing repeat bookings and understanding the reasons behind cancellations in City Hotels could also be beneficial.
