```python
final_df=pd.merge(df,df2,on='Country Code',how="left")
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df["Country"].value_counts()
```

```python
a=sns.countplot(x=final_df["Country"])
for bars in a.containers:
  a.bar_label(bars)
plt.rcParams["figure.figsize"]=(22,5)
```

```python
final_df.columns
```

```python
rating=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text'],as_index=False).size()
```

```python
rating.head()
```

```python
sns.barplot(x="Aggregate rating",y="size",data=rating,hue="Rating color",palette=["White","red","orange","yellow","green","DarkGreen"])
plt.rcParams["figure.figsize"]=(12,6)

```

```python
sns.countplot(x="Rating color",data=rating,palette=["White","red","orange","yellow","green","DarkGreen"])
```

```python
final_df[final_df['Aggregate rating']==0.0].groupby(["Country"]).size().reset_index().rename(columns={0:"count"}).sort_values(by="count",ascending=False)
```

```python
final_df[final_df['Aggregate rating']==0.0].groupby(["Country","Aggregate rating"]).size().reset_index().rename(columns={0:"count"}).sort_values(by="count",ascending=False)
```

```python
a=final_df[final_df['Has Online delivery']=="Yes"].groupby(["Country"]).size().reset_index().rename(columns={0:"count"}).sort_values(by="count",ascending=False)
```

```python
mk=sns.barplot(x="Country",y="count",data=a)
plt.rcParams["figure.figsize"]=(10,5)
for bars in mk.containers:
  mk.bar_label(bars)
```

```python
b=final_df[["Country","Has Online delivery"]].groupby(["Country"]).size().reset_index().rename(columns={0:"count"}).sort_values(by="count",ascending=False)
```

```python
mj=sns.barplot(x="Country",y="count",data=b)
plt.rcParams["figure.figsize"]=(25,5)
for bars in mj.containers:
  mj.bar_label(bars)
```

```python
a=final_df.groupby(["City"]).size().reset_index().rename(columns={0:"count"}).sort_values(by="count",ascending=False)[:5]
```

```python
mo=sns.barplot(x="City",y="count",data=a)
for bars in mo.containers:
  mo.bar_label(bars)
```
